```python
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic')
plt.rcParams['font.family'] ='NanumBarunGothic'
plt.rc('axes', unicode_minus=False)
plt.rcParams['axes.unicode_minus'] = False
```

```python

import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/경찰청_범죄 발생 지역별 통계_20221231.csv'
data = pd.read_csv(file_path, encoding='cp949')

seoul_data = data.loc[:, ['범죄대분류', '서울']]

crime_summary = seoul_data.groupby('범죄대분류').sum()

relevant_crimes = crime_summary.loc[['강력범죄', '절도범죄', '폭력범죄', '지능범죄']]

plt.figure(figsize=(8, 8))
plt.pie(relevant_crimes['서울'], labels=relevant_crimes.index, autopct='%1.1f%%', startangle=140)
plt.title('서울의 주요 범죄 비율')
plt.show()
```
